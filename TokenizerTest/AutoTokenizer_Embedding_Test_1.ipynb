## 추천시스템의 설계
1. 일차적으로 회사이름과 직무가 맞는 사람들을 단순 필터링 한다.(가장 우선순위)<br><br>
2. 일차에서 필터링 한 사람들을 대상으로 회사의 섹터, 전공은 코사인 유사도로 나머지 Label encoding된 변수들은 맞는 것들의 카운트를 통해서 우선순위를 정한다.<br><br>
3. 업데이트를 하게되면 그 다음 우선순위인 사람들을 추천하게 된다.

* 기본적으로 IBCF UBCF는 적합하지 않다.

```python
C = tokenizer.encode('에너지')
D = tokenizer.encode('소재')
if len(C)<len(D):
    C = pad_sequences([C],maxlen = len(D),padding = 'pre')
    C = C.reshape(1,-1)
    D = np.array(D).reshape(1,-1)
else:
    D = pad_sequences([D],maxlen = len(C), padding = 'pre')
    D = D.reshape(1,-1)
    C = np.array(C).reshape(1,-1)
cosine_similarity(C,D)
```
