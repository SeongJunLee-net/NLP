```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel, AdamW

# KcBERT 모델과 토크나이저 로드
model_name = "beomi/kcbert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name,output_hidden_states = True)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.metrics.pairwise import cosine_similarity
```

```python
C = tokenizer.encode('화학공학과')
D = tokenizer.encode('신소재공학과')
if len(C)<len(D):
    C = pad_sequences([C],maxlen = len(D),padding = 'post')
    C = C.reshape(-1)
    D = np.array(D).reshape(-1)
    token_ids = torch.tensor(C).unsqueeze(0)
    A = model(token_ids)
    token_ids = torch.tensor(D).unsqueeze(0)
    B = model(token_ids)
else:
    D = pad_sequences([D],maxlen = len(C), padding = 'post')
    D = D.reshape(-1)
    C = np.array(C).reshape(-1)
    token_ids = torch.tensor(C).unsqueeze(0)
    A = model(token_ids)
    token_ids = torch.tensor(D).unsqueeze(0)
    B = model(token_ids)
cosine_similarity(A['last_hidden_state'].view(1,-1).detach().numpy(),B['last_hidden_state'].view(1,-1).detach().numpy())
```
